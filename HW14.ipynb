### 1. [3 балла] Preprocessing

Возьмите себе копию "Мира":

Необходимо написать функцию, которая считывает текст из файла и с помощью pymystem3 лемматизирует его.

Аргумент функции - путь к файлу (с названием файла). Возвращаемое значение - лемматизированный текст.

```python
from pymystem3 import Mystem

def lemmatize(path):
    text = open(path, 'r')
    m = Mystem()
    lemmas = m.lemmatize(text.read())
    #print(''.join(lemmas))
    def removeall(x,y):
        while x in y: y.remove(x)
    removeall(' ',lemmas)
    removeall('\ufeff',lemmas)
    removeall('\n',lemmas)
    removeall('.',lemmas)
    removeall(',',lemmas)
    removeall(', ',lemmas)
    removeall(':',lemmas)
    removeall(': ',lemmas)
    removeall('. ',lemmas)
    for z in range(0,9):
        removeall(str(z),lemmas)
    removeall(' — ',lemmas)
    removeall('? ',lemmas)
    removeall(', — ',lemmas)
    removeall(' \n',lemmas)
    removeall(':\n',lemmas)
    removeall(' «',lemmas)
    removeall('» ',lemmas)
    removeall('»',lemmas)
    slem = ' '.join(lemmas)
    return slem
lemmatize('harms_mir.txt')
slems = lemmatize('harms_mir.txt')
slems
```
